# Ohm's Law

Ohm's law states that the current flowing through a conductor between two points is directly proportional to the voltage across the two points and inversely proportional to the resistance between them. Mathematically, it can be expressed as V = IR, where V is the voltage (in volts), I is the current (in amperes), and R is the resistance (in ohms).

```python
def ohms_law_calculator(V=None, I=None, R=None, P=None):
    # print(V,I,R,P)
    if V is not None and I is not None:
        R = V / I
        P = V * I
        # return R,P
    elif V is not None and R is not None:
        I = V / R
        P = V**2 / R
        # return I,P
    elif I is not None and R is not None:
        V = I * R
        P = I**2 * R
        # return V,P
    elif P is not None and V is not None:
        I = P / V
        R = V / I
        # return I,R
    elif P is not None and I is not None:
        V = P / I
        R = V / I
        # return V,R
    elif P is not None and R is not None:
        V = (P * R)**0.5
        I = V / R
        # return V,I
    else:
        return "Please provide at least two of the values (V, I, R, P)."
    return f"Voltage: {round(V,2)} V, Current: {round(I,2)} A, Resistance: {round(R,2)} Ω, Power: {round(P,2)} W"

print(ohms_law_calculator(V=120, I=1))
print(ohms_law_calculator(V=120, R=100))
print(ohms_law_calculator(I=1, R=100))
print(ohms_law_calculator(P=100, V=120))
print(ohms_law_calculator(P=100, I=1))
print(ohms_law_calculator(P=100, R=100))
```

### Plotting the relationship between Voltage & Current | Voltage & Power | Voltage & Current

```python
import matplotlib.pyplot as plt
import numpy as np

def ohms_law(V, I, R, P):
    R = V / I
    P = V * I
    return R, P

V = np.linspace(1, 10, 100) # start, stop, number of samples
I = np.linspace(1, 10, 100)
R, P = ohms_law(V, I, None, None)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
ax1.plot(V, R, label='Resistance')
ax1.set_xlabel('Voltage (V)')
ax1.set_ylabel('Resistance (Ω)')
ax1.legend()

ax2.plot(V, P, label='Power')
ax2.set_xlabel('Voltage (V)')
ax2.set_ylabel('Power (W)')
ax2.legend()

ax3.plot(V, I, label='Current')
ax3.set_xlabel('Voltage (V)')
ax3.set_ylabel('Current (I)')
ax3.legend()

plt.show()
```
