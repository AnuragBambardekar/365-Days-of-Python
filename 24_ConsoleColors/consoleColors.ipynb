# Console colors using ANSI codes

#### Basic colors
Difefrent consoles render these codes at different values
\x1b[30m	foreground black <br>
\x1b[31m	foreground red <br>
\x1b[32m	foreground green <br>
\x1b[33m	foreground yellow <br>
\x1b[34m	foreground blue <br>
\x1b[35m	foreground magenta <br>
\x1b[36m	foreground cyan <br>
\x1b[37m	foreground white <br>
\x1b[40m	background black <br>
\x1b[41m	background red <br>
\x1b[42m	background green <br>
\x1b[43m	background yellow <br>
\x1b[44m	background blue <br>
\x1b[45m	background magenta <br>
\x1b[46m	background cyan <br>
\x1b[47m	background white <br>

```python
print("\x1b[31mhello, I am Anurag")
print("\x1b[32mhello, I am Anurag")
print("\x1b[33mhello, I am Anurag")
print("\x1b[34mhello, I am Anurag")
print("\x1b[35mhello, I am Anurag")
print("\x1b[36mhello, I am Anurag")
print("\x1b[37mhello, I am Anurag")
print("\x1b[40mhello, I am Anurag")
print("\x1b[41mhello, I am Anurag")
print("\x1b[42mhello, I am Anurag")
print("\x1b[43mhello, I am Anurag")
print("\x1b[44mhello, I am Anurag")
print("\x1b[45mhello, I am Anurag")
print("\x1b[46mhello, I am Anurag")
print("\x1b[47m\x1b[31mhello, I am Anurag")
```

#### 256-colors
\x1b[38;5;[n]m is foreground <br>
\x1b[48;5;[n]m is background <br>

The first 16 entries are the basic color table again <br>

The next 216 entries are a 6x6x6 cube <br>

The final 24 entries are a grayscale ramp <br>

## Foreground

```python
for n in range(1,257):
    print(f"\x1b[38;5;{n}mhello, I am Anurag")
```

## Background

```python
for n in range(1,256):
    print(f"\x1b[48;5;{n}mhello, I am Anurag")
```

## 256 - Color mode

```python
for n in range(1,256):
    print(f"\x1b[38;2;0;0;{n}mhello, I am Anurag")
```
