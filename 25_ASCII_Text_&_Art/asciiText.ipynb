```python
import pyfiglet
```

```python
print(pyfiglet.figlet_format("hello"))
```

```python
print(pyfiglet.figlet_format("hello", justify="center"))
```

```python
print(pyfiglet.figlet_format("hello", justify="center", width=110))
```

### List of fonts supported by figlet

```python
fig = pyfiglet.Figlet()
fig.getFonts()
```

### Let's see what each font is

```python
for font in fig.getFonts():
    print(font)
    print(pyfiglet.figlet_format("hello", font=font))
```
