# Descriptive Statistics

The mean represents the average value of the dataset and is affected by outliers, whereas the median represents the middle value of the dataset and is not affected by outliers. The mode represents the most common value in the dataset. These three measures of central tendency are important because they provide different insights into the distribution of the data.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# np.random.seed(21782) # so that results are consistent on each run

# Generate some random data
data = np.random.normal(10, 2, 1000) # Draw random samples from a normal (Gaussian) distribution.

# Calculate the mean, median, and mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]


# Plot a histogram of the data
plt.hist(data, bins=50)

# Add vertical lines for the mean, median, and mode
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(median, color='g', linestyle='dashed', linewidth=1)
plt.axvline(mode, color='b', linestyle='dashed', linewidth=1)

# Add a legend
plt.legend(['Mean = {:.2f}'.format(mean),
            'Median = {:.2f}'.format(median),
            'Mode = {:.2f}'.format(mode)])

# Show the plot
plt.show()

# Some Analysis to understand the plot
if mean == median == mode:
    print("The data is perfectly symmetrical.")
elif mean == median:
    print("The data is skewed but balanced.")
elif median == mode:
    print("The data is skewed towards the mode.")
else:
    print("The data is skewed.")
```

Let's see a dataset which is skewed towards the mode

The majority of the data is generated around mode1 (80) using normal distribution with mean of 80 and standard deviation of 10. And the minority of the data is generated around mode2 (60) using normal distribution with mean of 60 and standard deviation of 10. Then the data is concatenated to create a single dataset.The resulting plot will show two peaks, but one peak is much higher than the other, which makes the data skewed towards the mode.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data with two peaks
mode1 = 80
mode2 = 60
size = 1000

data1 = np.random.normal(mode1, 10, int(size*0.8)) # majority of data around mode1
data2 = np.random.normal(mode2, 10, int(size*0.2)) # minority of data around mode2

data_new = np.concatenate([data1, data2])

# Plot the histogram of the data
plt.hist(data_new, bins=30, alpha=0.7, color='blue')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Bimodal Distribution Skewed Towards the Mode')
plt.axvline(x=mode1, color='y', label='Mode1')
plt.axvline(x=mode2, color='pink', label='Mode2')

# Calculate the mean, median, and mode
mean = np.mean(data_new)
median = np.median(data_new)

# Add vertical lines for the mean, median, and mode
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label='Median')

# Add a legend
plt.legend()

# Show the plot
plt.show()

if mean == median == mode1 == mode2:
    print("The data is perfectly symmetrical.")
elif mean == median:
    print("The data is skewed but balanced.")
elif round(median/10) * 10 == mode1:
    print("The data is skewed towards the mode1.")
elif round(median/10) * 10 == mode2:
    print("The data is skewed towards the mode2.")
else:
    print("The data is skewed towards one of the modes.")
```

A perfectly Symmetric case.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set the mode of the distribution
mode = 5

# Compute the mean and median of the distribution based on the mode
mean = mode
median = mode

# Create a normal distribution with mean and median equal to mode
# and standard deviation equal to 1
dist = norm(loc=mode, scale=1)

# Generate some random data from the distribution
data = dist.rvs(size=1000)

# Plot the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Normal Distribution with Mean = Median = Mode = ' + str(mode))
plt.axvline(x=mean, color='r', label='Mean')
plt.axvline(x=median, color='g', label='Median')
plt.axvline(x=mode, color='magenta', label='Mode')
plt.legend()

# Show the plot
plt.show()

if mean == median == mode:
    print("The data is perfectly symmetrical.")
elif mean == median:
    print("The data is skewed but balanced.")
elif round(median/10) * 10 == mode:
    print("The data is skewed towards the mode.")
else:
    print("The data is skewed.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

# Generate some sample data with a symmetric but not normal distribution
data = np.concatenate([np.random.normal(loc=0, scale=1, size=500),
                        np.random.normal(loc=4, scale=1, size=50),
                        np.random.normal(loc=-4, scale=1, size=50)])

# Compute mean and median of the data
mean = np.mean(data)
median = np.median(data)

# Create a histogram of the data
plt.hist(data, bins=20, density=True, alpha=0.5, color='blue')
plt.axvline(x=mean, color='red', label='Mean')
plt.axvline(x=median, color='green', label='Median')
plt.legend()
plt.show()

# Create a density plot of the data
plt.hist(data, bins=20, density=True, alpha=0.5, color='blue')
plt.plot(np.linspace(-8, 8, 1000), 
         scipy.stats.norm.pdf(np.linspace(-8, 8, 1000), loc=np.mean(data), scale=np.std(data)), 
         color='red', label='Normal distribution')
plt.legend()
plt.show()

if mean == median == mode:
    print("The data is perfectly symmetrical.")
elif round(mean/10) * 10 == round(median/10) * 10:
    print("The data is skewed but balanced.")
elif round(median/10) * 10 == mode:
    print("The data is skewed towards the mode.")
else:
    print("The data is skewed.")
```
