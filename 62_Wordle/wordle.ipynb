A more robust implementation, where it shows you on each attempt, the correctly/incorrectly guessed letters and in the end shows you the guessed words throughout the game.

```python
import random
import nltk
import keyboard

word_list = nltk.corpus.words.words()

# Filter the word list to include only 5-letter words
five_letter_words = [word.lower() for word in word_list if len(word) == 5]

def get_word():
    Word_ = [word for word in word_list if len(word) == 5]
    return random.choice(Word_)

def play_wordle(word_input):
    # Choose a random word from the list
    word = word_input.lower()

    # List to store the guessed letters and words
    guessed_incorrect_letters = []
    guessed_correct_letters = []
    guessed_words = []

    valid_guesses = 0

    # Loop to play the game
    for i in range(6):
        while True:
            # Prompt user to guess the word
            guess = input("Enter a five-letter word: ")

            # Check if the user pressed the Escape key
            if keyboard.is_pressed('Esc'):
                print("You exited the game.")
                return

            # Check if the guess has exactly 5 letters
            if len(guess) != 5:
                print("Please enter a five-letter word.")
                continue

            # Check if the word is valid
            if guess not in five_letter_words:
                print("Invalid word! Please enter a valid word.")
                continue

            # Check if the word has already been guessed
            if guess in guessed_words:
                print("You've already guessed that word!")
                continue

            # If everything is valid, break out of the loop
            guessed_words.append(guess)
            break

        valid_guesses += 1

        # Check if the guess is the same as the word
        if guess == word:
            print("You win!")
            break
        else:
            # Check which letters are correct
            correct = ""
            for j in range(5):
                if guess[j] == word[j]:
                    correct += guess[j]
                    guessed_correct_letters += guess[j]
                elif guess[j] != word[j]:
                    guessed_incorrect_letters += guess[j]
                    correct += "*"
                else:
                    correct += "*"

            # Print the correct letters and prompt for another guess
            print(f"{correct} ({i + 1}/6)")
            print(f"Guessed Incorrect letters: {', '.join(str(item) for item in set(guessed_incorrect_letters))}")
            print(f"Guessed Correct letters: {', '.join(str(item) for item in set(guessed_correct_letters))}")

    else:
        # Player has run out of guesses
        print(f"You lose! The word was {word.capitalize()}.")

    print(f"Number of valid guesses: {valid_guesses}")
    print(f"Guessed words: {', '.join(guessed_words)}")

word = get_word()
play_wordle(word)
```
