# BINGO

In the United States, bingo is a game of chance in which each player matches the numbers printed in different arrangements on cards. The game host (caller) draws at random, marking the selected numbers with tiles. When a player finds the selected numbers are arranged on their card in a row, they call out "Bingo!" to alert all participants to a winning card, which prompts the game host (or an associate assisting the host) to examine the card for verification of the win. Players compete against one another to be the first to have a winning arrangement for the prize or jackpot. After a winner is declared, the players clear their number cards of the tiles and the game host begins a new round of play. ~Wikipedia

```python
import random

# Create a list of numbers from 1 to 75
numbers = list(range(1, 76))

# Shuffle the numbers
random.shuffle(numbers)

# Create a bingo card with a random selection of 24 numbers
card = random.sample(numbers, 25)

# Add a free space to the center of the card
card[12] = "Free"

#Count number of attempts
count=0

# Print the bingo card
print("BINGO CARD")
print(" " + "-"*25)
for i in range(0, 24, 5):
    row = "|"
    for j in range(5):
        index = i + j
        if index == 12:
            row += "Free|"
        elif index < len(card):
            row += f" {card[index]:2} |"
        else:
            row += "        |"
    print(row+"")
    print(" " + "-"*25)

# Draw numbers and check for matches
drawn_numbers = []
while True:
    # Draw a number
    number = random.choice(numbers)
    # Check if the number has already been drawn
    if number not in drawn_numbers:
        # Mark the number on the card if it's a match
        if number in card:
            card[card.index(number)] = "X"
        # Print the number and the updated card
        print("Number drawn:", number)
        count+=1
        print(" " + "-"*25)
        for i in range(0, 24, 5):
            row = "|"
            for j in range(5):
                index = i + j
                if index == 12:
                    row += "Free|"
                elif index < len(card):
                    if card[index] == "X":
                        row += " X  |"
                    else:
                        row += f" {card[index]:2} |"
                else:
                    row += "        |"
            print(row)
            print(" " + "-"*25)
        # Check for a bingo
        if (card[0] == "X" and card[1] == "X" and card[2] == "X" and card[3] == "X" and card[4] == "X" or # Horizontal top row
            card[5] == "X" and card[6] == "X" and card[7] == "X" and card[8] == "X" and card[9] == "X" or # Horizontal middle row
            card[10] == "X" and card[11] == "X" and card[12] == "X" and card[13] == "X" and card[14] == "X" or # Horizontal bottom row
            card[0] == "X" and card[5] == "X" and card[10] == "X" and card[15] == "X" and card[20] == "X" or # Vertical left column
            card[1] == "X" and card[6] == "X" and card[11] == "X" and card[16] == "X" and card[21] == "X" or # Vertical middle column
            card[2] == "X" and card[7] == "X" and card[12] == "X" and card[17] == "X" and card[22] == "X" or # Vertical right column
            card[0] == "X" and card[6] == "X" and card[12] == "X" and card[18] == "X" and card[24] == "X" or # Diagonal top-left to bottom-right
            card[4] == "X" and card[8] == "X" and card[12] == "X" and card[16] == "X" and card[20] == "X"): # Diagonal top-right to bottom-left
            
            print("BINGO!")
            print("Attempts taken: ", count)
            break
        # Add the number to the list of drawn numbers
        drawn_numbers.append(number)
```
