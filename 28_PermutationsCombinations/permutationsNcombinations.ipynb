```python
import itertools
```

When order does not matter, we use combinations

```python
my_list = [1,2,3]
combinations = itertools.combinations(my_list, 3)
for c in combinations:
    print(c)
```

```python
combinations = itertools.combinations(my_list, 2)
for c in combinations:
    print(c)
```

When order does matter, we use permutations

```python
permutations = itertools.permutations(my_list, 2)
for c in permutations:
    print(c)
```

```python
string = "abc"
permutations = itertools.permutations(string)

for l in permutations:
    print(l)
```

### Find out all possible combinations that add up to 10

```python
my_list2 = [1,2,3,4,5,6]
combinations = itertools.combinations(my_list2, 3)
permutations = itertools.permutations(my_list2, 3)

#what values add upto 10
print([res for res in combinations if sum(res) == 10])
```

```python
print([res for res in permutations if sum(res) == 10])
```

It's better to use combinations in the above situation

Let's see a situation where Permutations are appropriate

```python
word = 'sample'
my_letters = 'plmeas'

combinations = itertools.combinations(my_letters, 6)
permutations = itertools.permutations(my_letters, 6)

for p in combinations:
    print(p)
```

```python
for p in permutations:
    print(p)
```

```python
for p in permutations:
    print(p)
    if ''.join(p) == word:
        print('Match')
        break
    else:
        print("No Match")
```
