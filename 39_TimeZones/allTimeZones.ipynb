# Timezones

```python
import pytz
import datetime

for tz in pytz.all_timezones:
    current_time = datetime.datetime.now(pytz.timezone(tz))
    print(f"Current time in {tz}: {current_time.strftime('%Y-%m-%d %H:%M:%S %Z%z')}")
```

### Current Time in India

```python
import pytz
import datetime

india_timezone = pytz.timezone("Asia/Kolkata")
current_time = datetime.datetime.now(india_timezone)

print(f"Current time in India: {current_time.strftime('%Y-%m-%d %H:%M:%S %Z%z')}")
```
