# Performance of Searching Algorithms

```python
import time

# Linear Search Algorithm
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

# Test the Algorithms
arr = [i for i in range(1000000)]
x = 999999

start_time = time.time()
linear_search(arr, x)
end_time = time.time()
print(f"Linear Search took {end_time - start_time} seconds.")
print(f"Elapsed time: {(end_time - start_time) * 1000:.3f} milliseconds")


```

## Visualising O(n)

Code visualizes the growth as the problem instance (N) increases of several functions: log n, n, n^{3}, e^{n}

```python
import numpy as np
import matplotlib.pyplot as plt
  
# Returns a function that computes x ^ n for a given n
def poly(n):
    def polyXN(x):
        return x**n
    return polyXN
  
# Functions to compare and colors to use in the graph
FUNCTIONS = [np.log, poly(1), poly(2), poly(3), np.exp]
COLORS = ['c', 'b', 'm', 'y', 'r']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the graphs 
def compareAsymptotic(ax, n):
    x = np.arange(1, n, 1)
    ax.set_title('O(n) for n =' + str(n))
    for f, c in zip(FUNCTIONS, COLORS):
        ax.plot(x, f(x), c)

# Call the function for each subplot
compareAsymptotic(axs[0, 0], 3)
compareAsymptotic(axs[0, 1], 5)
compareAsymptotic(axs[1, 0], 10)
compareAsymptotic(axs[1, 1], 20)

# Set overall title for the entire plot
plt.suptitle('Comparing Asymptotic Behaviour of Several Functions')

# Display the plot
plt.show()
```

The exponential function has the worst performance, as it grows very quick given the input size. For N=20, the other functions are insignificant compared with the exponential.

The logarithm is shown in cyan, the polynomials in blue, magenta and yellow and the exponential in red.

### Similarly, for O(n^2)

```python
import numpy as np
import matplotlib.pyplot as plt
  
# Returns a function that computes x ^ n for a given n
def quad():
    def quadX(x):
        return x**2
    return quadX
  
# Functions to compare and colors to use in the graph
FUNCTIONS = [np.log, lambda x: x, quad(), lambda x: x*np.log(x), lambda x: x**3]
COLORS = ['c', 'b', 'm', 'y', 'r']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the graphs 
def compareAsymptotic(ax, n):
    x = np.arange(1, n, 1)
    ax.set_title('O(n^2) for n =' + str(n))
    for f, c in zip(FUNCTIONS, COLORS):
        ax.plot(x, f(x), c)

# Call the function for each subplot
compareAsymptotic(axs[0, 0], 10)
compareAsymptotic(axs[0, 1], 20)
compareAsymptotic(axs[1, 0], 50)
compareAsymptotic(axs[1, 1], 100)

# Set overall title for the entire plot
plt.suptitle('Comparing Asymptotic Behaviour of Several Functions')

# Display the plot
plt.show()
```
