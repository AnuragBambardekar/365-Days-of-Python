# Blood Pressure Calculator

Systolic and diastolic blood pressure are two measurements of the pressure of the blood in your arteries.

Systolic blood pressure is the top number in a blood pressure reading and measures the pressure in your arteries when your heart beats or contracts to pump blood out. Diastolic blood pressure is the bottom number in a blood pressure reading and measures the pressure in your arteries when your heart is at rest or between beats.

The unit of measurement for blood pressure is millimeters of mercury (mmHg), and blood pressure is typically expressed as systolic/diastolic, for example, 120/80 mmHg.

**Note that these values are just examples and are not intended to represent normal or healthy blood pressure ranges.**

```python
import matplotlib.pyplot as plt

bp_readings = []

for i in range(3):
    systolic = int(input("Enter systolic blood pressure: "))
    diastolic = int(input("Enter diastolic blood pressure: "))
    bp_readings.append((systolic, diastolic))

# Create lists of systolic and diastolic readings for plotting
systolic_readings = [i[0] for i in bp_readings]
diastolic_readings = [i[1] for i in bp_readings]

systolic_values = [120, 130, 135, 125, 140, 150]
diastolic_values = [80, 85, 90, 75, 95, 100]

ideal_systolic = [120] * len(systolic_values)  # create a list of the same length as systolic_values with ideal systolic value
ideal_diastolic = [80] * len(diastolic_values)  # create a list of the same length as diastolic_values with ideal diastolic value

plt.plot(systolic_values, label='Systolic')
plt.plot(diastolic_values, label='Diastolic')
plt.plot(ideal_systolic, '--', label='Ideal Systolic')  # plot the ideal systolic line as a dashed line
plt.plot(ideal_diastolic, '--', label='Ideal Diastolic')  # plot the ideal diastolic line as a dashed line
plt.legend()
plt.title('Blood Pressure')
plt.xlabel('Reading')
plt.ylabel('mmHg')
plt.show()

# Plot the systolic and diastolic readings on a line graph
plt.plot(systolic_readings, label='Systolic')
plt.plot(diastolic_readings, label='Diastolic')
plt.xlabel('Reading Number')
plt.ylabel('Blood Pressure')
plt.title('Blood Pressure Readings Over Time')
plt.legend()
plt.show()

average_systolic = sum(systolic_readings) / len(systolic_readings)
average_diastolic = sum(diastolic_readings) / len(diastolic_readings)

print("Average systolic blood pressure:", average_systolic, "mmHg")
print("Average diastolic blood pressure:", average_diastolic, "mmHg")
```
