```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 5000) 
# -np.pi/2 is the starting value of the interval
# np.pi/2 is the ending value of the interval
# 10 is the number of evenly spaced values that should be generated within the interval
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.figure(1)
plt.subplot(311) # 3 rows, 1 column, 1st index
plt.plot(x, y_sin)
plt.title('Sine Function', x=1.2, y=0.5)
plt.grid(True)

plt.subplot(312) # 3 rows, 1 columns, 2nd index
plt.plot(x, y_cos)
plt.title('Cosine Function', x=1.2, y=0.5)
plt.grid(True)

plt.subplot(313) # 3 rows, 1 columns, 3rd index
plt.plot(x, y_tan)
plt.title('Tangent Function', x=1.2, y=0.5)
plt.xlabel('x')
plt.grid(True)

plt.savefig('trigo_function.svg')

plt.subplots_adjust(hspace=1.4)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi/2, np.pi/2, 300)
y_csc = 1/np.sin(x)
y_sec = 1/np.cos(x)
y_cot = 1/np.tan(x)

plt.figure(1,figsize=(8, 6))
plt.subplot(311)
plt.plot(x, y_csc, label='Cosecant')
plt.title('Cosecant Function', x=1.2, y=0.5)
plt.grid(True)

plt.subplot(312)
plt.plot(x, y_sec, label='Secant')
plt.title('Secant Function', x=1.2, y=0.5)
plt.grid(True)

plt.subplot(313)
plt.plot(x, y_cot, label='Cotangent')
plt.title('Cotangent Function', x=1.2, y=0.5)
plt.xlabel('x')
plt.grid(True)

plt.savefig('trigo2_function.svg')

plt.subplots_adjust(hspace=1.4)
plt.show()
```
