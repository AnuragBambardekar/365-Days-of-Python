# FizzBuzz - Algorithm Analysis [Time and Space Complexity]

A program that prints the numbers from 1 to 100. But for multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

## Algorithm Analysis

The time complexity of this FizzBuzz program is O(n), where n is the number of iterations (i.e., the range of numbers from 1 to 100 in this case). This is because the program is using a simple loop to iterate over the numbers and the number of iterations is proportional to n. <br>

The space complexity of this program is also O(n), because the program only stores a constant amount of information for each iteration (i.e., the strings "Fizz", "Buzz", and "FizzBuzz" or the integer value of the number itself), and the amount of memory used is proportional to the number of iterations n. <br>

```python
fizz = 3
buzz = 5
for i in range(1, 101):
    output = ""
    fizz -= 1
    buzz -= 1
    if fizz == 0:
        output += "Fizz"
        fizz = 3
    if buzz == 0:
        output += "Buzz"
        buzz = 5
    if output == "":
        output = str(i)
    print(output)
```

This approach has a space complexity of O(1), because it only needs to store a constant amount of information for each iteration (i.e., the output string), regardless of the size of the input. The time complexity is also O(n), since we still need to iterate through all the numbers from 1 to n. However, the performance of this implementation may be better for larger values of n, since addition can be faster than the modulo operation for some values.

A simple variation where I replace 3 with Fizz and 5 with Buzz.

```python
def fizz_buzz(n):
    for i in range(1, n+1):
        num_str = str(i)
        output = ""
        for digit in num_str:
            if digit == "3":
                output += "Fizz"
            elif digit == "5":
                output += "Buzz"
            else:
                output += digit
        print(output)
fizz_buzz(100)
```
