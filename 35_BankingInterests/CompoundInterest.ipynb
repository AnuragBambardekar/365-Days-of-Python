# Compound Interest

```python
import math
def compound_interest(principal, rate, time, n):
    #print(principal)
    #print(rate)
    #print(time)
    #print(n)
    frequency = {
        "annually": 1,
        "semi-annually": 2,
        "quarterly": 4,
        "bi-monthly": 6,
        "monthly": 12,
        "semi-monthly": 24,
        "bi-weekly": 26,
        "weekly": 52,
        "daily (365)": 365,
        "daily (360)": 360,
        "continuously": None
    }

    c_freq = frequency[n]
    #print(c_freq)
    #amt = principal * (1 + rate / (c_freq * 100)) ** (c_freq * time)
    #print(amt)
    if c_freq is None:
        return principal * (math.e) ** (rate * time / 100)
    return principal * (1 + rate / (c_freq * 100)) ** (c_freq * time)

p = float(input("Enter the principal amount: "))
r = float(input("Enter the rate of interest: "))
t = float(input("Enter the time (in years): "))
print("\nCompounding Frequency Options:")
print("1. Annually")
print("2. Semi-Annually")
print("3. Quarterly")
print("4. Bi-Monthly")
print("5. Monthly")
print("6. Semi-Monthly")
print("7. Bi-Weekly")
print("8. Weekly")
print("9. Daily (365)")
print("10. Daily (360)")
print("11. Continuously")

option = input("\nEnter the number corresponding to the desired compounding frequency: ")
frequency_options = {
    "1": "annually",
    "2": "semi-annually",
    "3": "quarterly",
    "4": "bi-monthly",
    "5": "monthly",
    "6": "semi-monthly",
    "7": "bi-weekly",
    "8": "weekly",
    "9": "daily (365)",
    "10": "daily (360)",
    "11": "continuously"
}

n = frequency_options[option]
# print(n)

# Convert decimal years to equivalent number of years
t = int(t) + (t % 1) * 365 / 365

ci = compound_interest(p, r, t, n)

print("=============================")
print("Principal amount: ", p)
print("Rate of Interest : ", r)
print("Time : ", t, " years")
if option == "1":
    print("Compounding Frequency : Annually")
elif option == "2":
    print("Compounding Frequency : Semi-annually")
elif option == "3":
    print("Compounding Frequency : Quarterly")
elif option == "4":
    print("Compounding Frequency : Bi-monthly")
elif option == "5":
    print("Compounding Frequency : Monthly")
elif option == "6":
    print("Compounding Frequency : Semi-monthly")
elif option == "7":
    print("Compounding Frequency : Bi-weekly")
elif option == "8":
    print("Compounding Frequency : Weekly")
elif option == "9":
    print("Compounding Frequency : Daily (365)")
elif option == "10":
    print("Compounding Frequency : Daily (360)")
elif option == "11":
    print("Compounding Frequency : Continuously")
print("=============================")
print("Final Amount with Compound Interest: ", ci)
```
